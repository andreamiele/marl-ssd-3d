```python
from collections import defaultdict

df_dict = defaultdict(dict)

for ray_degree in [45, 85, 125]:
    for ray_length in [20, 30, 40]:
        df = pd.read_csv(f"rd_{ray_degree}_rl_{ray_length}.csv")
        df_dict[str(ray_degree)][str(ray_length)] = df
```

```python
df_dict['45'].keys()
```

```python
ray_length_list = []
ray_degree_list = []
lwr_list = []

for ray_degree, length_dict in df_dict.items():
    for ray_length, df in length_dict.items():
        ray_length_list.append(ray_length)
        ray_degree_list.append(ray_degree)
        lwr_list.append(df["lone_wolf_rate"].iloc[-1].item())

```

```python
df_lwr = pd.DataFrame({"ray_length":ray_length_list,"ray_degree":ray_degree_list,"lwr":lwr_list})
```

```python
import seaborn as sns
```

```python
df_lwr_pivot = df_lwr.pivot_table(index="ray_length", columns="ray_degree", values="lwr")
```

```python
sns.heatmap(df_lwr_pivot,annot=True,cmap="Blues")
plt.show()
```
