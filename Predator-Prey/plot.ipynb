```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from PIL import Image

df = pd.read_csv("inferences/rd_45_rl_20.csv")
df_captures = df[df.capture_x != -1]

bg_image = Image.open("background_map_flat.jpeg")

height = 50
width = 50
map_array = np.zeros((height,width))

x_min = 2.37
x_max = 69.72

z_min = 1.79
z_max = 69.08

x_offset = x_max - x_min
z_offset = z_max - z_min

for _, row in df_captures.iterrows():
    
    x_coord = int(height * ((row["capture_x"] - x_min) / x_offset))
    z_coord = int(width * ((-row["capture_z"] - z_min) / z_offset))
    map_array[x_coord,z_coord] += 1

# Display the background image
plt.imshow(bg_image, extent=[0, map_array.shape[1], map_array.shape[0], 0])

# Overlay the heatmap with transparency (alpha)
sns.heatmap(np.log(map_array.transpose()+1), cmap="Reds", alpha=0.6, cbar=False)

# Remove axis ticks
plt.xticks([])
plt.yticks([])

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from PIL import Image
import imageio
import io

# Load data
df = pd.read_csv("inferences/rd_45_rl_20.csv")
df_captures = df[df.capture_x != -1]

# Load background image
bg_image = Image.open("background_map_flat.jpeg")

# Constants for map size
height = 50
width = 50
map_array = np.zeros((height, width))

# Coordinate bounds
x_min = 2.37
x_max = 69.72
z_min = 1.79
z_max = 69.08
x_offset = x_max - x_min
z_offset = z_max - z_min

# Store frames
frames = []

for i, (_, row) in enumerate(df_captures.iterrows()):
    # Compute grid indices
    x_coord = int(height * ((row["capture_x"] - x_min) / x_offset))
    z_coord = int(width * ((-row["capture_z"] - z_min) / z_offset))
    map_array[x_coord, z_coord] += 1

    # Create a figure and overlay the heatmap on background
    fig, ax = plt.subplots()
    ax.imshow(bg_image, extent=[0, width, height, 0])
    sns.heatmap(
        np.log(map_array.transpose() + 1),
        cmap="Reds",
        alpha=0.6,
        cbar=False,
        ax=ax
    )
    ax.axis('off')

    # Save frame to buffer
    buf = io.BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight', pad_inches=0)
    buf.seek(0)
    frames.append(imageio.v2.imread(buf))
    plt.close(fig)

# Save frames as a GIF
imageio.mimsave("heatmap_evolution.gif", frames, duration=0.2)
```
